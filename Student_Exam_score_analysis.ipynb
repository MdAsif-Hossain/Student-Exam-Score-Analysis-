```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_scores.csv")
print(df.head())

```

```python
df.info()
```

```python
df.isnull().sum()
# total count of null values 
```

# Drop unnamed calumn

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

# Gender Distribution

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x ="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

### From the avove chart we have analysed that the number of females in the data is more than males

# Analysis Based on Perents Education

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize = (5,5))

sns.heatmap(gb, annot =True)
plt.title("Relationship between Parent's Education amd Student's Score")
plt.show()
```

### from the avob chart we have concluded that the education of parents have a good impact on student's result

# Analisis Based on Parents Matital Status

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize = (5,5))
sns.heatmap(gb1, annot =True)
plt.title("Relationship between Parent's Marital Status amd Student's Score")
plt.show()
```

### Parents marital status have almost no impact

# Distribution of Ethnic Group

```python
print(df["EthnicGroup"].unique())
```

```python

groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(mlist,labels = l, autopct = "%1.2f%%")
plt.title("Ethnic Distribution")
plt.show()
```
